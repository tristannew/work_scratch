```python
import pandas as pd
import requests
import pprint
```

```python
sdg_list = requests.get("https://unstats.un.org/sdgs/UNSDGAPIV5/v1/sdg/Goal/List?includechildren=true")
```

```python
print(sdg_list)
```

```python
sdg_goals_dict = {}
for item in sdg_list.json():
    for dict in item['targets']:
        for final in dict['indicators']:
            sdg_goals_dict[final['code']] = final['description']
```

```python
print(len(sdg_goals_dict))
```
